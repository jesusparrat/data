# Limpiar Dataframe

```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
```

```python
# Leemos el fichero CSV, especificando que el separador es ';'
df = pd.read_csv('datos.csv', sep = ';')
```

```python
# Mostramos el Dataframe
df
```

```python
# Eliminamos las columnas 'Nacional' y 'Sexo'
df = df.drop(['Nacional','Sexo'], axis = 1)
```

```python
# Mostramos el Dataframe
df
```

```python
# Renomabramos las columnas
df.rename(columns = {'Comunidades y Ciudades Autónomas':'Comunidad', 'Total':'Poblacion'}, inplace = True)
```

```python
# Mostramos el Dataframe
df
```

```python
type(df['Comunidad'])
```

```python
# Eliminamos el código numérico de la columna 'Comunidad'
df['Comunidad'] = df['Comunidad'].str[3:]
```

```python
# Mostramos el Dataframe
df
```

```python
# Eliminamos el separador de millares de la columna 'Población'
df['Poblacion'] = df['Poblacion'].str.replace('.', '', regex=False)
```

```python
# Mostramos el Dataframe
df
```

```python
# Guardamos el Dataframe limpio en formato CSV
df.to_csv('datos_limpios.csv')
```
