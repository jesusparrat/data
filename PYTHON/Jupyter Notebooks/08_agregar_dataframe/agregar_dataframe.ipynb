# Agregar una columna a un Dataframe

```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
```

```python
# Leemos el fichero CSV asignando los tipos deseados a las columnas
df = pd.read_csv(
    'datos_limpios.csv',
    index_col=[0],
    dtype={'Comunidad':'string', 'Periodo':'int16', 'Poblacion':'int32'}
)
```

```python
# Mostramos el Dataframe
df
```

## Agregar una columna

```python
# Agregamos la columna 'País' con el valor 'España'
df['Pais'] = 'España'
```

```python
# Mostramos el Dataframe
df
```

```python
# Guardamos el Dataframe en formato CSV
df.to_csv('datos_pais.csv')
```
