# Ficheros CSV

Cuando abrimos un fichero CSV podemos necesitar configurar dos parámetros:

- Codificación de caracteres
- Carácter separador

Antes de comenzar importaremos las librerías necesarias:

```python
# Importamos las librerías necesarias
import pandas as pd
```

```python
# Podemos obtener la codificación de caracteres del fichero
with open('precipitaciones_provincias_mm.csv') as f:
    file_encoding = f.encoding
```

```python
# Mostramos la codificación de caracteres del fichero
file_encoding
```

```python
# Al crear el Dataframe podemos especificar qué
# columna contiene los índices.
df = pd.read_csv('precipitaciones_provincias_mm.csv', encoding = file_encoding, sep = ';')
```

```python
# Mostramos el Dataframe
df
```
