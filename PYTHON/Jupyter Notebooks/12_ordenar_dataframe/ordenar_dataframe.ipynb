# Ordenar un Dataframe

```python
import pandas as pd
```

```python
# Creamos el Dataframe a partir del dataset en CSV
df = pd.read_csv('titanic.csv', index_col = [0])
```

```python
# Mostramos el Dataframe
df
```

```python
# Podemos ordenar el Dataframe por una columna dada con el
# método sort_values().
# Por ejemplo, para ordenarlo por Tarifa:
df.sort_values('Fare')
```

```python
df
```

```python
# Por defecto la ordenación se llevará a cabo en sentido ascendente,
# pero si queremos podemos especificar lo contrario
df.sort_values('Fare', ascending = False)
```

```python
# No obstante, es posible actualizar el Dataframe original si se desea,
# con el argumento inplace = True:
df.sort_values('Fare', ascending = False, inplace = True)
```

```python
# Mostramos el Dataframe, que ahora sí estará ordenado
df
```

```python
# La ordenación del Dataframe se puede llevar a cabo por cualquier columna
# cuyo tipo mantenga una relación de orden.
# Por ejemplo, podríamos llevar a cabo una ordenación alfabética por la columna
# 'Name':
df.sort_values(['Name'], inplace = True)
```

```python
# Mostramos el Dataframe
df
```

```python
# También podemos ordenar un Dataframe por varios campos.
# Por ejemplo, podemos ordenarlo primero por apellido, y
# seguidamente por nombre.
# Para ello dividimos primero la columna 'Name' en 'Name' y 'Surname'
df[['Surname','Name']] = df['Name'].str.split(pat = ',', expand = True)
```

```python
# Mostramos el Dataframe
df
```

```python
# A continuación eliminamos los tratamientos 'Mr.', 'Mrs.', etc.
df['Name'] = df['Name'].replace(['Miss ','Mr\. ','Ms\. ', 'Mrs\. '],'', regex = True)
```

```python
# Mostramos el Dataframe
df
```

```python
# Seguidamente ordenamos por apellido y nombre
df.sort_values(['Surname','Name'])
```

```python
# Mostramos el Dataframe
df
```

```python
# Si queremos volver a ordenar el Dataframe por su index podemos
# hacerlo con el método sort_index()
df.sort_index(inplace=True)
```

```python
# Mostramos el Dataframe
df
```
