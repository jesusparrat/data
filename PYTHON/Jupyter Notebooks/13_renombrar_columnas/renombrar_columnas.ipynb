# Renombrar columnas

```python
import pandas as pd
```

```python
# Creamos el Dataframe a partir del dataset en CSV
df = pd.read_csv('titanic.csv', index_col = [0])
```

```python
# Mostramos el Dataframe
df
```

```python
# Podemos renombrar columnas con el mÃ©todo rename()
df.rename(columns = {
    'Sex':'Gender'
}, inplace = True)
```

```python
# Mostramos el Dataframe
df
```

```python
# Podemos renombrar tanto el Ã­ndice como los nombres de las columnas
# con los atributos index y columns
df.index.names=['Index']
df.columns.names=['Fields']
```

```python
# Mostramos el Dataframe
df
```
